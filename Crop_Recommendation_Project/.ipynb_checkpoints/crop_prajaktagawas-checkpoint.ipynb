# Encoding the object columns using label encoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df["label"].unique()
```

```python
df["label"] = encoder.fit_transform(df["label"])
y = df.label
y
```

```python
sns.heatmap(df.corr(),annot=True)
```

# Model Bulding
Preparing data for model building

```python
X = df[["N","P","K","temperature","humidity","ph","rainfall"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# classification model building using LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# classification model building using NaiveBaye

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# Predicting the test values
Predicting the output for the test data

```python
pred =  model.predict(X_test)
pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
```

```python
LAB= ['rice', 'maize', 'chickpea', 'kidneybeans', 'pigeonpeas',
       'mothbeans', 'mungbean', 'blackgram', 'lentil', 'pomegranate',
       'banana', 'mango', 'watermelon', 'muskmelon', 'orange', 'papaya',
       'coconut', 'cotton', 'jute', 'coffee']
sns.heatmap(cm,annot=True,fmt="d",xticklabels=LAB,yticklabels=LAB)
```

# GaussianNB accuracy is greater than LogisticRegression

```python
prediction= model.predict([[90,42,43,21,82,7,203]])
prediction
```
