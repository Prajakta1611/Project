```python
Max_length = pickle.load(open('Max_length.pkl', 'rb'))
```

```python
Input_Seq = pickle.load(open('Input_Seq.pkl', 'rb'))
```

```python
# input_text = "Mr Modi could have pushed"
input_text = input("enter a text :")
predict_next_words= int(input("how many word you want to predict after the sentence ?"))

for i in range(predict_next_words):
    token_list = MyToken.texts_to_sequences([input_text])[0]
    print(token_list)
    token_list = pad_sequences([token_list], maxlen=Max_length-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=-1)
    output_word = ""
    for word, index in MyToken.word_index.items():
        if index == predicted:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```
