# Data Cleaning Model bulding
Identifying the column datatypes and Droping the unneccesary columns

```python
df.drop(["Unnamed: 0","Batsman"],axis=1,inplace=True)
```

```python
for i in df:
    print(i," : ",df[i].unique())
```

```python
df = df.replace("-",0)
```

```python
df["Bat1"].unique()
```

```python
df["Bat1"] =df["Bat1"].replace({"DNB":0,"absent":0,"TDNB":0,"sub":0})
```

```python
df["Bat1"] =df["Bat1"].apply(lambda x : str(x))
```

```python
df["Bat1"] =df["Bat1"].apply(lambda x : x.split("*"))
```

```python
df["Bat1"] =df["Bat1"].apply(lambda x : x[0])
```

```python
df["Bat1"] =df["Bat1"].apply(lambda x : int(x))
```

# Encoding the object columns 

```python
final = pd.get_dummies(data=df,drop_first=True,dtype=int)
final
```

```python
final.info()
```

# Model Bulding
Preparing data for model building

```python
X = final.drop("Runs",axis=1)
```

```python
y = final.Runs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```


# Regression model using LinearRegression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# predicting the value

```python
pred =  model.predict(X_test)
```
